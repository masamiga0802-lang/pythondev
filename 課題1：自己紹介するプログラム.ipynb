# 自己紹介するプログラム

このノートブックでは、名前と年齢を入力してもらい、来年の年齢を表示するプログラムを作成します。

## 1. 名前を入力する

```python
# 命令1: 名前を入力してもらい、変数 name に格納する
name = input("名前を入力してください：")
print(f"入力された名前: {name}")
```

## 2. 年齢を入力する

```python
# 命令2: 現在の年齢を入力してもらい、変数 age に格納する（整数に変換）
age = int(input("現在の年齢を入力してください："))
print(f"入力された年齢: {age} 歳")
```

## 3. 自己紹介を表示する

```python
# 命令3: 名前と来年の年齢を表示する
print(f"私の名前は {name} です。来年で {age + 1} 歳になります。")
```
