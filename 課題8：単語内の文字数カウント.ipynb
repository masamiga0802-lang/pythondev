```python
# 空のリストを作成
words = []

# アルファベット小文字すべてをキーに、値0の辞書を作成
char_count = {chr(i): 0 for i in range(ord('a'), ord('z') + 1)}

# 英単語の入力を繰り返す
while True:
    word = input("英単語を入力してください：")
    if word == "":
        break
    words.append(word)
    for char in word:
        if char in char_count:
            char_count[char] += 1

# ABC順に並べ替えて表示
print(f"入力した英単語：{sorted(words)}")

# 1文字以上あった文字を表示
for char, count in char_count.items():
    if count > 0:
        print(f"{char}が{count}個ありました")
```
